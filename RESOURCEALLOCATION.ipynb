#### From the above barplot, it is seen that 75% of the 9,601 households in North Horr, a subcounty in Marsabit county use the open/ bush as their human waste disposal method. This can be attributed to lack of basic or enough information about how human waste should be handled from medical professionals. As seen from the health facilities plot, Marsabit county ranks amongst the last counties with the lowest or least number of health facilities. This is one basis of poor community sensitization as with low number of health facilities arises a shortage of health professionals ranging from clinical officers to doctors.

# <center><font color='red'>CONCLUSION</font></center>
### From the above analysis, the following points are clear:
* There is unequal medical resource allocation in Kenya and this affects rural areas and The counties of Norther Kenya.
* The available have less oprational beds thus posing a dangerous picture moreso in Nothern Kenya.
* Most Hospitals or health facilities in kenya are owned by the ministry of health. This thus calls onto different investors that would like to invest in the Kenyan medical sector as the clear picture evidenced by the above analysis shows that enough has not been done in order to caution this sector interms of enough operational hospitals and hospital beds.
* Of all the health facilities in Kenya, almost 50% of them are community based health centeres and of these facilities, 34% are either semi-operational or not oprating at all.

#### Narrowing down the analysis to sub-county level, this analysis looks into the real situation of North Horr Sub-county in Marsabit county and the following is noted:
* Human waste disposal: It is noted from this analysis that most of the households in North Horr sub-county lack a basic proper way of disposing off their human waste. The analysis shows that over 70% of these households dispose their waste into the open fields or bush which is harmful not only to them but also the wellbeing of their livestock.
* Drinking water: This analysis further looks into the main source of drinking water for the residents of North Horr. It is noted that over 57% of these households actually get their drinking water from unprotected wells. This water is harmful and not clean enough for human consumption.

#### In order to understand if the above problems in the sub-county are as a result of lack of enough information, this analysis goes further to analyse the populations highest level of education completed, portion of the population possessing a mobile phone, a laptop or desktop and finally the number of people connected or using the internet. The followng was noted:
* Firstly, it was noted that less than 500 houselds in this sub-county have atleast one person that has completed university education. Ironically, it was noted that over 4000 houselholds have atleast one person completing primary school education. Also, over 1500 of these households, have atleast one person completing high school education. It is good to note that almost 1500 of these households have atleast one person not attending school at all. This therefore sends a warning shock and should be a matter of high concern in that the government and all involved stakeholders should put more effort in trying to understand why there is a high education disparity in this sub-county.
* Secondly, this study noted that of the 62,863 residents of this sub-county, only 14,024 possess a mobile phone translating to 22.3% of the entire North Horr population. Of the 22.3%, 56% are male and 44% female.
* Thirdly, this study noted that the level of computer illiteracy in this sub-county is high. This is because, only 1307 residents out of the 62,863 total population of the sub-county have access to either a laptop or a desktop. Of these, 882 are male and 424 female.
* Lastly, this study notes that only 3,071 residents of North Horr have access to the internet. Of these, 1983 are male and 1087 female.

#### <font color='purple'>This therefore, presents a sub-county that lacks not only enough medical facilities and resources but also one that lacks access to enough information both from its locals, nationally and internationally.</font>
